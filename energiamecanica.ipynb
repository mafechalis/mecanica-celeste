```python
import numpy as np
import matplotlib.pyplot as plt
import pymcel as pc
import plotly.graph_objects as go
```

```python
sistema=[
    dict(m=1,r=[0,0,0],v=[0,0,0]),
    dict(m=1,r=[-1,0,0],v=[0,1,0.5]),
    dict(m=0.1,r=[1,0,0],v=[0,-0.1,0])
]

ts=np.linspace(0,5,50)
rs,vs,rps,vps,cuadraturas= pc.ncuerpos_solucion(sistema,ts) 
```

```python
import plotly.graph_objects as go
import numpy as np

# Generar datos de ejemplo: posiciones de partículas en el tiempo
np.random.seed(42)
n_particulas = 10
n_frames = 100

# Crear trayectorias de partículas (movimiento aleatorio)
x_data = np.cumsum(np.random.randn(n_frames, n_particulas), axis=0)
y_data = np.cumsum(np.random.randn(n_frames, n_particulas), axis=0)
z_data = np.cumsum(np.random.randn(n_frames, n_particulas), axis=0)

# Crear frames para la animación
frames = []
for frame_idx in range(n_frames):
    frame = go.Frame(
        data=[go.Scatter3d(
            x=x_data[frame_idx],
            y=y_data[frame_idx],
            z=z_data[frame_idx],
            mode='markers',
            marker=dict(
                size=8,
                color=np.arange(n_particulas),
                colorscale='Viridis',
                showscale=True,
                colorbar=dict(title="Partícula")
            ),
            text=[f"Partícula {i}" for i in range(n_particulas)],
            hovertemplate="<b>%{text}</b><br>X: %{x:.2f}<br>Y: %{y:.2f}<br>Z: %{z:.2f}<extra></extra>"
        )],
        name=f"Frame {frame_idx}"
    )
    frames.append(frame)

# Crear figura inicial
fig = go.Figure(
    data=[go.Scatter3d(
        x=x_data[0],
        y=y_data[0],
        z=z_data[0],
        mode='markers',
        marker=dict(
            size=8,
            color=np.arange(n_particulas),
            colorscale='Viridis',
            showscale=True,
            colorbar=dict(title="Partícula")
        ),
        text=[f"Partícula {i}" for i in range(n_particulas)],
        hovertemplate="<b>%{text}</b><br>X: %{x:.2f}<br>Y: %{y:.2f}<br>Z: %{z:.2f}<extra></extra>"
    )],
    frames=frames
)

# Configurar controles de animación
fig.update_layout(
    title="Animación 3D de Partículas",
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z",
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.5)
        )
    ),
    updatemenus=[
        dict(
            type="buttons",
            showactive=False,
            buttons=[
                dict(
                    label="▶ Reproducir",
                    method="animate",
                    args=[None, {
                        "frame": {"duration": 50, "redraw": True},
                        "fromcurrent": True,
                        "transition": {"duration": 0}
                    }]
                ),
                dict(
                    label="⏸ Pausa",
                    method="animate",
                    args=[[None], {
                        "frame": {"duration": 0, "redraw": False},
                        "mode": "immediate",
                        "transition": {"duration": 0}
                    }]
                )
            ]
        )
    ],
    sliders=[dict(
        active=0,
        steps=[
            dict(
                args=[[f.name], {
                    "frame": {"duration": 0, "redraw": True},
                    "mode": "immediate",
                    "transition": {"duration": 0}
                }],
                method="animate",
                label=f"{i}"
            )
            for i, f in enumerate(frames)
        ]
    )],
    height=700,
    width=900
)

fig.show()
```

```python
G = 1  # Constante gravitacional
U = 0

for i in range(len(sistema)):
    for j in range(i + 1, len(sistema)):
        # Calcular la distancia entre las partículas i y j
        r_ij = np.sqrt(sum((np.array(sistema[i]['r']) - np.array(sistema[j]['r']))**2))
        # Sumar la contribución de la energía potencial
        U += -G * sistema[i]['m'] * sistema[j]['m'] / r_ij

U
```
